# 层次聚类

## 执行聚类

```r
# 没安装flexclust包的需要先安装
data(nutrient, package = "flexclust")
head(nutrient)
```

```r
row.names(nutrient) <- tolower(row.names(nutrient))
head(nutrient)
```

```r
dim(nutrient)
```

```r
# 聚类前先标准化
nutrient_scaled <- scale(nutrient)
head(nutrient_scaled)
```

```r
# 然后计算距离，方法就选择欧氏距离。这也是默认方法
d <- dist(nutrient_scaled, method = "euclidean")
```

```r
h_clust <- hclust(d, method = "average")
```

```r
# 提取3类
nutrient_hclust <- cutree(h_clust, k = 3)
```

```r
plot(h_clust)
```

```r
h_clust <- hclust(d, method = "complete")
plot(h_clust)
```

## 确定参数

```r
library(NbClust)
nc <- NbClust(
  nutrient_scaled,
  min.nc = 2,
  max.nc = 15,
  distance = "euclidean",
  method = "average"
)
```

```r
plot(h_clust, hang = -1, main = "", xlab = "")
rect.hclust(h_clust, k = 5)
```

# K聚类

```r
data(wine, package = "rattle")
# 标准化数据
df <- scale(wine[,-1])

psych::headTail(df)
```

```r
library(NbClust)

nc <- NbClust(
  df,
  min.nc = 2,
  max.nc = 15,
  method = "kmeans"
)
```

```r
table(nc$Best.nc[1,])
## 
##  0  1  2  3 14 15 
##  2  1  2 19  1  1

barplot(table(nc$Best.nc[1,]),
        xlab = "聚类数目",
        ylab = "评判准则个数"
        )
```

```r
library(factoextra)
## Loading required package: ggplot2
## Welcome! Want to learn more? See two factoextra-related books at https://goo.gl/ve3WBa

set.seed(123)
fviz_nbclust(df, kmeans, k.max = 15)
```

```r
# 设置种子数保证结果可复现
set.seed(123)
fit.km <- kmeans(df, centers = 3, nstart = 25) # 尝试多种配置选择最好的一个
fit.km
```

```r
fviz_cluster(fit.km, data = df)
```

```r
fviz_cluster(fit.km, data = df, 
             ellipse = T, # 增加椭圆
             ellipse.type = "t", # 椭圆类型
             geom = "point", # 只显示点不要文字
             palette = "lancet", # 支持超多配色方案
             ggtheme = theme_bw() # 支持更换主题
             )
```

```r
library(flexclust)
## Loading required package: grid
## Loading required package: lattice
## Loading required package: modeltools
## Loading required package: stats4

tb <- table(wine$Type, fit.km$cluster)
randIndex(tb)
```

# PAM

```r
library(cluster)

set.seed(123)
fit.pam <- pam(wine[,-1], k=3 # 聚为3类
               , stand = T # 聚类前进行标准化
               )
fit.pam
```

```r
clusplot(fit.pam, main = "PAM cluster")
```

```r
fviz_cluster(fit.pam, 
             ellipse = T, # 增加椭圆
             ellipse.type = "t", # 椭圆类型
             geom = "point", # 只显示点不要文字
             palette = "aaas", # 支持超多配色方案
             ggtheme = theme_bw() # 支持更换主题
             )
```

```r
tb <- table(wine$Type, fit.pam$clustering)
randIndex(tb)
```
