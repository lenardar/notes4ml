# 测试层次聚类

```python
# 导入必要的库
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
```

```python
# 导入数据
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
X
```

```python
y = iris.target
y
```

```python
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
model.fit(X)
```

```python
model.labels_
```

```python
# 绘制原始散点图
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[y == 0, 0], X[y == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y == 1, 0], X[y == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y == 2, 0], X[y == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.title('Real clusters')

# 绘制预测散点图
plt.subplot(1, 2, 2)
plt.scatter(X[model.labels_ == 0, 0], X[model.labels_ == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[model.labels_ == 1, 0], X[model.labels_ == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[model.labels_ == 2, 0], X[model.labels_ == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.title('Clusters of customers')
plt.legend()
```

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)
    
    from scipy.cluster.hierarchy import dendrogram
    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
# 绘制树状图
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plot_dendrogram(model, labels=model.labels_)
plt.show()
```

绘制主成分分析散点图

```python
import scipy.cluster.hierarchy as sch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel('./层次聚类/data_cluster2.xlsx')

# 将索引变为city列
data.index = data['city']
data = data.drop('city',axis=1)
data
```

```python
#不同的位置裁剪即可得到不同的聚类数目
label = sch.cut_tree(model, height=400)

data_copy = data.copy()
data_copy['height_400'] = label
data_copy
```

```python
# 训练模型
from sklearn.decomposition import PCA

pca = PCA(n_components = 0.95)    #选择方差累积占比95%的主成分
pca.fit(data)

data_pca = pca.transform(data)
data_pca
```

```python
# 绘制散点图
plt.figure(figsize=(10, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=data_copy['height_400'])
for i in range(data_pca.shape[0]):
    plt.text(data_pca[i, 0], data_pca[i, 1], data.index[i])
    
x_label = 'PC1(%s%%)' % round((pca.explained_variance_ratio_[0]*100.0),2)   #x轴标签字符串
y_label = 'PC2(%s%%)' % round((pca.explained_variance_ratio_[1]*100.0),2)   #y轴标签字符串
plt.xlabel(x_label)    #绘制x轴标签
plt.ylabel(y_label)    #绘制y轴标签
```
