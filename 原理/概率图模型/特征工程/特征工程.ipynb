# 数据预处理

## 归一化

### 测试归一化-MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np

data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
pd.DataFrame(data)
```

```python
# 归一化
scaler = MinMaxScaler()  # 实例化
scaler = scaler.fit(data)  # fit，在这里本质是生成min(x)和max(x)
result = scaler.transform(data)  # 通过接口导出结果
result
```

### 复原归一化

```python
# 复原归一化
result = scaler.inverse_transform(result)
result
```

```python
# 设置feature_range
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
scaler = MinMaxScaler(feature_range=(5, 10))  # 生成5-10之间的数据
result = scaler.fit_transform(data)
result
```

### 大量特征的归一化

```python
# 当X中的特征数量非常多的时候，fit会报错并表示，数据量太大了我计算不了
# 此时使用partial_fit作为训练接口
scaler = MinMaxScaler()
scaler = scaler.partial_fit(data)
result = scaler.transform(data)
result
```

### 使用numpy实现归一化

```python
# 使用numpy实现归一化
data = np.array(data)

min = data.min(axis=0)
max = data.max(axis=0)

data_nor = (data - min) / (max - min)
data_nor
```

## 标准化-StandardScaler

## 缺失值处理

### SimpleImputer

```python
import pandas as pd
data = pd.read_csv('data.csv', index_col=0)
data
```

```python
data.info()
```

```python
# 填补年龄
Age = data.loc[:, 'Age'].values.reshape(-1, 1)
Age[:20]
```

```python
# 填补缺失值
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer()  # 实例化，默认均值填补
imp_median = SimpleImputer(strategy='median')  # 用中位数填补
imp_0 = SimpleImputer(strategy='constant', fill_value=0)  # 用0填补
```

```python
imp_mean.fit_transform(Age)
```

```python
imp_median.fit_transform(Age)
```

```python
imp_0.fit_transform(Age)
```

```python
# 使用中位数填补Age
data.loc[:, 'Age'] = imp_median.fit_transform(data.loc[:, ['Age']])
data.info()
```

```python
# 使用众数填补Embarked
Embarked = data.loc[:, 'Embarked'].values.reshape(-1, 1)
imp_mode = SimpleImputer(strategy='most_frequent')
data.loc[:, 'Embarked'] = imp_mode.fit_transform(Embarked)
data.info()
```

## 处理连续性特征

### 二值化

```python
data_2 = data_.copy()

from sklearn.preprocessing import Binarizer

data_2.loc[:, 'Age'] = data_2.loc[:, 'Age'].fillna(data_2.loc[:, 'Age'].median())
data_2
```

```python
X = data_2.loc[:, 'Age'].values.reshape(-1, 1)
X
```

```python
# 二值化
binarizer = Binarizer(threshold=30).fit(X)  # 二值化的阈值为30
binarizer.transform(X)
```

### 分段

```python
from sklearn.preprocessing import KBinsDiscretizer

X = data_2.loc[:, 'Fare'].values.reshape(-1, 1)
X
```

```python
# 分箱
kbd = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')
result = kbd.fit_transform(X)
result
```
