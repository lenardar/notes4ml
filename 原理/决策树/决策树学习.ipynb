# 使用sklearn学习决策树

```python
from sklearn import tree  # 导入决策树模块
from sklearn.datasets import load_wine  # 导入红酒数据集
from sklearn.model_selection import train_test_split  # 导入数据集切分模块
import pandas as pd
```

```python
# 加载红酒数据集
wine = load_wine()
wine
```

```python
wine.data
```

```python
wine.target
```

```python
wine.target_names
```

# 对数据集进行分割，分成训练集和测试集

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data, wine.target, test_size=0.3)
```

```python
Xtrain.shape
```

```python
wine.data.shape
```

```python
Ytrain.shape
```

```python
Ytrain
```

# 建立决策树模型

```python
clf = tree.DecisionTreeClassifier(criterion="entropy")
clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)

score
```

# 树的可视化

我是真的服了，直接调用graphviz包死活用不了，最后只能用pygraphviz曲线救国了；<br>
然后这里又有一个新的可视化坑，可视化默认不支持中文，直接用会显示乱码，要在`export_graphviz`函数中添加`special_characters=True`、`fontname="Microsoft YaHei`，来解决中文显示的问题；<br>
正常notebook图片应该显示在下方，但这个包里不支持这个方法，真的麻烦！！！！ <br>
为了在下方显示图片，只能在from IPython.display import Image，麻烦爆了！！！！

```python
feature_name = ['酒精', '苹果酸', '灰', '灰的碱性', '镁', '总酚', '类黄酮',
                '非黄烷类酚类', '花青素', '颜色强度', '色调', 'od280/od315稀释葡萄酒', '脯氨酸']

import pygraphviz
from IPython.display import Image


dot_data = tree.export_graphviz(clf,
                                feature_names=feature_name,
                                class_names=["琴酒", "雪莉", "贝尔摩德"],
                                filled=True,
                                rounded=True,
                                special_characters=True, 
                                fontname="Microsoft YaHei" )

graph = pygraphviz.AGraph(dot_data)
graph.draw(prog='dot')
Image(graph.draw(format='png', prog='dot'))
```

## 绘制学习曲线

```python
import matplotlib.pyplot as plt

tets = []
for i in range(20):
    # 建立深度不同的决策树
    clf = tree.DecisionTreeClassifier(criterion="entropy",
                                      random_state=30,
                                      splitter="random",
                                      max_depth=i+1)
    clf.fit(Xtrain, Ytrain)
    score = clf.score(Xtest, Ytest)
    tets.append(score)
plt.plot(range(1, 21), tets, color="red", label="max_depth")
plt.legend()
plt.show()
```
