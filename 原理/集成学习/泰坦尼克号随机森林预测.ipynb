# 泰坦尼克号预测

## 导入必要的包和数据集

```python
# 导入必要的包
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 导入数据集data.csv
data = pd.read_csv("data.csv")
data
```

```python
X = data.iloc[:, data.columns != "Survived"]
X
```

```python
y = data.iloc[:, data.columns == "Survived"]
y
```

## 数据预处理

```python
X.info()
```

```python
# 删除Name列与Cabin列、PassengerId、Ticket	列
X = X.drop(["Name", "Cabin", "PassengerId", "Ticket"], axis=1)
X
```

```python
# 用平均值填充Age列的缺失值
X["Age"] = X["Age"].fillna(X["Age"].mean())
X.info()
```

```python
# 删除Embarked列的缺失值
X = X.dropna()
X.info()
```

```python
# 调整y
y = y.loc[X.index]
y
```

```python
# 重建索引
X.index = range(X.shape[0])
y.index = range(y.shape[0])
X
```

```python
X.describe()
```

```python
# 合并SibSp和Parch列
X["Family"] = X["SibSp"] + X["Parch"]
X = X.drop(["SibSp", "Parch"], axis=1)

# 如果Family列的值大于0，将其设置为1
X["Family"] = X["Family"].map(lambda x: 1 if x > 0 else 0)
X
```

```python
Embarked_index = list(X['Embarked'].unique())
Embarked_index
```

```python
# 将Embarked列的值转换为数值
X["Embarked"] = X["Embarked"].map(lambda x: Embarked_index.index(x))
X
```

```python
# 将Sex列的值转换为数值
X['Sex'] = X['Sex'].map(lambda x: ['male', 'female'].index(x))
X
```

```python
X.info()
```

## 构建模型

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rfc = RandomForestClassifier(n_estimators=100, random_state=90)
score_pre = cross_val_score(rfc, X, y.values.ravel(), cv=10).mean()
score_pre
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rfc = RandomForestClassifier(n_estimators=500, random_state=90)
score_pre = cross_val_score(rfc, X, y.values.ravel(), cv=10).mean()
score_pre
```

```python
# 对n_estimators进行调参
scorel = []

for i in range(0, 200, 10):
    rfc = RandomForestClassifier(n_estimators=i + 1, n_jobs=-1, random_state=90)
    score = cross_val_score(rfc, X, y.values.ravel(), cv=10).mean()
    scorel.append(score)
    
print(max(scorel), (scorel.index(max(scorel)) * 10) + 1)
plt.figure(figsize=[20, 5])
plt.plot(range(1, 201, 10), scorel)
plt.show()
```

```python
# 对n_estimators进一步微调
scorel = []

for i in range(140, 160):
    rfc = RandomForestClassifier(n_estimators=i, n_jobs=-1, random_state=90)
    score = cross_val_score(rfc, X, y.values.ravel(), cv=10).mean()
    scorel.append(score)
    
print(max(scorel), ([*range(140, 160)][scorel.index(max(scorel))]))
plt.figure(figsize=[20, 5])
plt.plot(range(140, 160), scorel)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'max_depth': np.arange(1, 20, 1)}

# 对max_depth进行调参
rfc = RandomForestClassifier(n_estimators=151, random_state=90)

GS = GridSearchCV(rfc, param_grid, cv=10)
GS.fit(X, y.values.ravel())
```
